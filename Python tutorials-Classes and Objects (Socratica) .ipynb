```python
# classes group together data (fields) and functions (methods)

class User:
    pass # pass does nothing, you need it here because atleast one line is necessary to define a class

user1 = User() # user1 is instance or object of User class
user1.first_name = "Maria" # first_name and last_name are fields, they are used to hold data for single object of a class
user1.last_name = "Shaukat"
```

```python
print(user1.first_name) # see you can attach fields to an object without defining it in the class
# but its safer to define it
```

```python
import datetime

class Userr:
    """A member of FF. This class stores data on each member"""
    def __init__(self, full_name, birthday):
        self.name = full_name # assigning objects to fields
        self.birthday = birthday #yyyymmdd
        
        name_pieces = full_name.split(' ')
        self.first_name = name_pieces[0]
        self.last_name = name_pieces[-1]
        
    def age(self):
        """Returns the age of Userr object in years"""
        today = datetime.datetime.today()
        bdate = datetime.datetime.strptime(self.birthday, '%Y%m%d')
        return today.year - bdate.year
```
